```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/Users/aarsh/Desktop/CGVI/Thesis/Data/Test_3840x1920/0008/image/000002.jpg', cv2.IMREAD_COLOR)
# Initiate ORB detector
orb = cv2.ORB_create()
 
# find the keypoints with ORB
kp = orb.detect(image,None)
 
# compute the descriptors with ORB
kp, des = orb.compute(image, kp)

plt.figure(figsize=(20, 20))
 
# draw only keypoints location,not size and orientation
img2 = cv2.drawKeypoints(image, kp, None, color=(0,255,0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img2[:, :, ::-1])
plt.show()
```
